**3. ONE HOT ENCODING USING SKLEARN**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:4], df.iloc[:,-1],
                                                    test_size = 0.2,
                                                    random_state = 2)
```

```python
ohe = OneHotEncoder(drop='first', sparse = False, dtype = np.int32) #implentation without column transformer to define pipeline
```

```python
X_train_new = ohe.fit_transform(X_train[['fuel','owner']])
```

```python
X_test_new = ohe.transform(X_test[['fuel','owner']])
```

**4. ONE HOT ENCODING WITH TOP CATEGORIES**

```python
counts = df['brand'].value_counts()
```

```python
df['brand'].nunique()
threshold = 100
```

```python
repl = counts[counts <= threshold].index
```

```python
pd.get_dummies(df['brand'].replace(repl, 'uncommon'))
```
