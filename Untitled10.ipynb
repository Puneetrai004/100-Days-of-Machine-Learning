# Predicting Solar Cycle Amplitude Using Decline Rate

## Objective
To analyze historical sunspot cycle data and establish a correlation between the decline rate of a solar cycle and its subsequent peak amplitude. This correlation is then used to predict the amplitude of the upcoming solar cycle.

## Methodology
1. **Data Acquisition:**
   - Sunspot number data was obtained from the Sunspot Index and Long-term Solar Observations (SILSO) database : https://www.sidc.be/SILSO/INFO/snytotcsv.php
   - Data cleaning was performed to remove any negative values or anomalies.

2. **Plotting Sunspot Data:**
   - A time-series plot of sunspot numbers (1700–present) was created to visualize cyclic variations.

3. **Peak and Trough Identification:**
   - Used the `find_peaks` function from SciPy to detect solar cycle maxima and minima based on prominence filtering.
   
4. **Calculating Decline Rates:**
   - The decline rate between consecutive cycle minima was computed using numerical differentiation.

5. **Linear Regression Analysis:**
   - Established a linear correlation between decline rates and peak amplitudes of subsequent cycles using `sklearn`'s `LinearRegression`.
   - The regression equation obtained was:
     - **Slope:** -23.2587
     - **Intercept:** 169.6608
   - Model performance metrics:
     - **R-squared (R²):** 0.0845 (indicating a weak correlation)
     - **RMSE:** 50.7752

6. **Prediction of Next Solar Cycle Amplitude:**
   - The most recent decline rate was **-0.0545**.
   - Using the regression model, the predicted amplitude of the next solar cycle was **170.93**.

## Results & Observations
- The regression model showed a weak correlation between the decline rate and the peak amplitude of the next cycle.
- Despite the low R² value, the model was used to estimate the next cycle’s peak sunspot number.
- A scatter plot with a regression line demonstrated the relationship visually, along with confidence intervals.


## Conclusion
- The decline rate of a solar cycle provides some predictive insight into the amplitude of the next cycle, though the correlation is weak.
- The model suggests a predicted peak sunspot number of **170.93** for the next cycle.
- Future improvements could involve non-linear models or incorporating additional solar parameters to enhance predictive accuracy.

**Necesaary Imports**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

**Load and clean data**

```python
url = 'https://www.sidc.be/SILSO/INFO/snytotcsv.php'
data = pd.read_csv(url, delimiter=';', header=None, names=['Year', 'SunspotNumber', 'Error', 'Observations', 'Extra'])
data = data.drop(columns=['Extra'])
data['Year'] = data['Year'].astype(float)
data['SunspotNumber'] = data['SunspotNumber'].astype(float)
data = data[data['SunspotNumber'] >= 0]
```

**Plot the raw sunspot data**

```python

plt.figure(figsize=(14, 6))
plt.plot(data['Year'], data['SunspotNumber'], 'b-', linewidth=1)
plt.title('Historical Sunspot Number (1700-Present)')
plt.xlabel('Year')
plt.ylabel('Sunspot Number')
plt.grid(True, alpha=0.3)
plt.savefig('sunspot_history.png', dpi=300, bbox_inches='tight')
plt.show()
```

**Identify peaks (maxima) and troughs (minima)**

```python
peaks, _ = find_peaks(data['SunspotNumber'], prominence=50)  # Increased prominence for clearer peaks
troughs, _ = find_peaks(-data['SunspotNumber'], prominence=50)
```

**Extract peak and trough values**

```python
peak_values = data.iloc[peaks]
trough_values = data.iloc[troughs]
```

**Plot with identified maxima and minima**

```python
plt.figure(figsize=(14, 6))
plt.plot(data['Year'], data['SunspotNumber'], 'b-', linewidth=1)
plt.plot(peak_values['Year'], peak_values['SunspotNumber'], 'ro', label='Cycle Maxima')
plt.plot(trough_values['Year'], trough_values['SunspotNumber'], 'go', label='Cycle Minima')
plt.title('Solar Cycle Maxima and Minima')
plt.xlabel('Year')
plt.ylabel('Sunspot Number')
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('solar_cycles.png', dpi=300, bbox_inches='tight')
plt.show()
```

**Calculate decline rates**

```python
decline_rates = []
min_years = []
for i in range(len(trough_values) - 1):
    decline_rate = (trough_values.iloc[i + 1]['SunspotNumber'] - trough_values.iloc[i]['SunspotNumber']) / \
                   (trough_values.iloc[i + 1]['Year'] - trough_values.iloc[i]['Year'])
    decline_rates.append(decline_rate)
    min_years.append(trough_values.iloc[i]['Year'])
```

**Prepare correlation data**

```python
correlation_data = pd.DataFrame({
    'MinYear': min_years,
    'PeakValue': peak_values['SunspotNumber'].values[:len(decline_rates)],
    'DeclineRate': decline_rates
})
```

**Display correlation data**

```python
print("Correlation Data:")
print(correlation_data)
```

**Scatter plot of PeakValue vs. DeclineRate**

```python
# Drop NaN values (if any)
correlation_data.dropna(inplace=True)

print("Correlation Data:")
print(correlation_data)

plt.figure(figsize=(8, 5))
plt.scatter(correlation_data['PeakValue'], correlation_data['DeclineRate'], color='b', alpha=0.7)
plt.xlabel('Peak Sunspot Number')
plt.ylabel('Decline Rate')
plt.title('Correlation between Sunspot Peak and Decline Rate')
plt.grid(True)
plt.show()
```

**Perform linear regression using sklearn**

```python
X = correlation_data['DeclineRate'].values.reshape(-1, 1)
y = correlation_data['PeakValue'].values
model = LinearRegression()
model.fit(X, y)
```

**Predict values and calculate Evaluation metrics**

```python
x_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_pred_range = model.predict(x_range)

predicted_values = model.predict(X)
r_squared = r2_score(y, predicted_values)
rmse = np.sqrt(mean_squared_error(y, predicted_values))
```

**Plot results with visualization**

```python
plt.figure(figsize=(12, 8))
plt.scatter(correlation_data['DeclineRate'], correlation_data['PeakValue'], label='Data Points', color='blue', alpha=0.7)
plt.plot(x_range, y_pred_range, color='red', linewidth=2, label=f'Best Fit Line (y = {model.coef_[0]:.2f}x + {model.intercept_:.2f})')

# Added confidence intervals
plt.fill_between(x_range.flatten(),
                 y_pred_range.flatten() - rmse,
                 y_pred_range.flatten() + rmse,
                 color='red', alpha=0.1, label=f'±RMSE ({rmse:.2f})')

plt.xlabel('Decline Rate', fontsize=12)
plt.ylabel('Peak Value', fontsize=12)
plt.title('Linear Correlation between Decline Rate and Peak Value', fontsize=14)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)

# Equation Annotation and R²
equation_text = f"y = {model.coef_[0]:.2f}x + {model.intercept_:.2f}"
r2_text = f"R² = {r_squared:.4f}"
plt.annotate(f"{equation_text}\n{r2_text}",
             xy=(0.05, 0.95),
             xycoords='axes fraction',
             bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))

plt.tight_layout()
plt.show()
```

**Result and predicting the amplitude of next cycle**


```python
# Display results
print(f"Slope: {model.coef_[0]:.4f}")
print(f"Intercept: {model.intercept_:.4f}")
print(f"R-squared: {r_squared:.4f}")
print(f"RMSE: {rmse:.4f}")

# Predict next cycle amplitude based on recent decline rate
recent_decline_rate = correlation_data['DeclineRate'].iloc[-1]
predicted_next_amplitude = model.predict([[recent_decline_rate]])[0]
print(f"\nMost recent decline rate: {recent_decline_rate:.4f}")
print(f"Predicted next cycle amplitude: {predicted_next_amplitude:.2f}")
```
